```python
from math import atan2, degrees

value_x = float(input("Give a value for x: "))
value_y = float(input("Give a value for y: "))

angle_r = atan2(value_y, value_x)
angle_d = degrees(angle_r)

# Convert negative angles to 0â€“360 range
if angle_d < 0:
    angle_d += 360

print("Angle:", angle_d)

s = "(x,y) belongs to quandrant "

if 0 <= angle_d < 90:
    quadrant = 1
elif 90 <= angle_d < 180:
    quadrant = 2
elif 180 <= angle_d < 270:
    quadrant = 3
else:
    quadrant = 4
    
print(s, quadrant)
```

math.atan just calculates the arctangent of one number (a division) in radians, which ranges from -pi/2 to pi/2. atan2() calculates the angle between a vector of (y,x)(the input is two numbers) and the positive x-axis, which ranges from -pi to pi. Due to this, atan2() can actually determine in which quadrant the input belongs.

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

#Print night1 and night2 as sets
n1 = set(night1)
print(f"Set version of night1; {n1}")
n2 = set(night2)
print(f"Set version of night2; {n2}")

inters = set.intersection(n1, n2)
common_values = list(inters) #Intersection values as a list
print(f"The common values of night1 and night2 are: {common_values}")

if 5 in common_values:
    print("The number 5 is in the intersection list")
else:
    print("The number 5 is not in the intersection list")

if 5 in n1:
    print("The number 5 is in the set version of night1")
else:
    print("The number 5 is not in the set version of night1")
    
```
