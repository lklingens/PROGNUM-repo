```python
import numpy as np

mu = 0.0 #mean
sigma = 1.4 #standard deviation

#Samples from the distribution
s = np.random.normal(mu, sigma, 100000) 

#Calculate and print standard deviation of the data
st_dev = np.std(s)
print(f"The standard deviation is: {st_dev}")
mean = np.mean(s)
print(mean)
```
