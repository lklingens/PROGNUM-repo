#### <span style="color:red"> *Question 4.3*

```python
import numpy as np

RPS = input("Enter R, P or S to start playing:")
possible_input = ['R','P','S']

# Text that is printed when the game is over.

Win = "You win, congratulations!"
Lose = "You lost."

# Generates a random number between 0 and 3, then gets the value of that number from the list possible_input

indx = np.random.randint(0,len(possible_input))
result = possible_input[indx]

def RockPapSci(RPS, result):
    """ This function evaluates a random value against user input in Rock-paper-Scissors, then prints if you've won or lost.
        This function written by Klingens on 03-03-2026.

    Parameters
    ----------
    RPS (str) the user input of R, P or S
    result (str) the random vale R, P or S, generated by the programme
    
    Return
    -------
    count (int) the number of times you tried to win
    """
    game = True  # condition
    count = 1
    while game:
        print(f"The programme chose: {result}")
        if RPS == 'R' and result == 'S':
            game = False  # Ends the game
            print(Win)
        elif RPS == 'P' and result == 'S':
            game = False
            print(Lose)
        elif RPS == 'R' and result == 'P':
            game = False
            print(Lose)
        elif RPS == 'P' and result == 'R':
            game = False
            print(Win)
        elif RPS == 'S' and result == 'P':
            game = False
            print(Win)
        elif RPS == 'S' and result == 'R':
            game = False
            print(Lose)
        elif RPS == result:
                print(f"It's a tie, try again.")
                RPS = input("Enter R, P or S to try again:")  # Asks user to enter a new number due to a draw.
                result = possible_input[indx]
                count += 1  # Ups the count of number of tries by 1
        elif RPS != 'R' or 'P' or 'S':  # Evaluates whether the input is valid. If not, enter something again.
            print("Invalid input, try again")
            RPS = input("Enter R, P or S to try again:")
    return count

print(f"You tried {RockPapSci(RPS, result)} times.")
```

#### <span style="color:red"> *Question 4.4*

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

# Create the data.

lengths = np.random.normal(185,8, 100000)
print(f"Set of random lengths: {lengths[:10]}")

# Calculate the mean and standard deviation.

mean = np.mean(lengths)
stdev = np.std(lengths)
print(f"The mean is: {mean} and the standard deviation is {stdev}")

# Probability of finding a value smaller or equal to the indicated height.

ind_height = float(input("Enter a height in cm:"))  # indicated height h
cdf = norm.cdf(ind_height, mean, stdev)  # If you do (h, stdev, mean), it is going to find heights larger than X.
P = cdf*100        # to percentage
print(f"Probability P(X ≤ {ind_height}) is: {P:.2f} %")

# Probability of a height located within 1 standard deviation.

stdev_1 = norm.cdf(mean+stdev, mean, stdev) - norm.cdf(mean-stdev, mean, stdev)
print(f"Probability P(within 1 standard deviation) is: {stdev_1:}")

# Plot the histrogram, mean and standard deviation.

plt.hist(lengths, bins=40, label='Normal distribution')
plt.axvline(mean, color='navy', label=f"Mean = {mean:.2f} cm")
plt.axvspan(mean-stdev, mean+stdev, color='darkturquoise', alpha=0.3, label=f"Standard deviation = {stdev:.2f} cm")

plt.title("Distribution of human lengths")
plt.xlabel("Height ($cm$)")
plt.ylabel("Number of people")

plt.legend()
plt.show()
```

#### <span style="color:red"> *Question 4.12*

```python
import numpy as np
import math
from matplotlib import pyplot as plt

angle_deg = 30  # degrees
theta = math.radians(angle_deg)  # angle in radians
rotor = math.cos(theta)+1j*math.sin(theta)  # rotor

# x and y coordinates

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

z = x + 1j*y  # Converts x and y coordinates to complex numbers

# Multiplying the complex coordinates by the rotor creates the rotated coordinates
# The x coordinates are the real part and the y coordinates are the imaginary numbers

x_rotor = (z*rotor).real
y_rotor = (z*rotor).imag

plt.plot(x, y, color='b',label='Original square')
plt.plot(x_rotor, y_rotor, color='r', label='Rotated square')
plt.title(f"Rotating a square over {angle_deg} degrees")

plt.gca().set_aspect(1)  # Scales the plot.
plt.legend()
plt.show()
```

#### <span style="color:red"> *Question 4.16*

```python
from astropy.cosmology import Planck18
from math import sqrt
from scipy import integrate

H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0

Mpc_to_km = 3.0857e19  # 1 Mpc = 3.0857e19 km
Gyr = 1e9  # in yr
yr = 3.15576e7  # in sec
Gyr_in_s = Gyr*yr # in sec

z = float(input("Enter your redshift:"))

def function(z):
    """A function that calculates the lookback time to redshift z""" 
    H = H0*sqrt(Om0*((1+z)**3)+Ode0)  # Units km/s/Mpc
    t = 1/((1+z)*H)  # 1/H, so Mpc*s/km
    t_in_Gyr = (t*Mpc_to_km)/Gyr_in_s  # Conversion to Gyr
    return t_in_Gyr  # in Gyr

time, error = integrate.quad(function, 0, z)  # Integration to get the time

print(f"The lookback time to redshift z = {z} is {time} Gyr, with an error of {error} Gyr.")
```

```python
from scipy.stats import norm 
import numpy as np
from matplotlib import pyplot as plt

# Generate data

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

# Axis for the PDF's

x_axis = np.linspace(min(X.min(), Y.min()), max(X.max(), Y.max()), 100000)  # Correct range for the x-axis of the PDF's

# Generate the PDF's

X_PDF = norm.pdf(x_axis, 0, 1)
Y_PDF = norm.pdf(x_axis, 2, 1)

# Plot the histrograms with some visual parameters.

plt.hist(X, bins=30, color='b', edgecolor='black', alpha=0.5, label='X', density=True)  # Density=True normalizes the histrogram to area=1
plt.hist(Y, bins=30, color='r', edgecolor='black', alpha=0.5, label='Y', density=True)

# Plot the PDF's

plt.plot(x_axis, X_PDF, color='k', linewidth=4, linestyle=':', label='X(PDF)')
plt.plot(x_axis, Y_PDF, color='k', linewidth=4, linestyle='--', label='Y(PDF)')


plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('1D Histrogram + PDF')
plt.grid(axis='y', linestyle='--')
plt.legend()
plt.show()
```
