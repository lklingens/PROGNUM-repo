#### *Question 4.1*

```python
X = [1,2,3,10]
Y =[10,20,30,40]
z = []

for x, y in zip(X,Y):
    z.append(x+y)

print(f"The sum of X and Y is: {z}")

Z = [x+y for x, y in zip(X,Y)]  # One liner
print(f"The sum of X and Y, using a one liner, is: {Z}")

def add(x,y):
    """ A function that add the values of x and y"""
    return x+y
    
sums = map(add, X, Y)  # Implement the function on the two lists to add their values
print(f"The type of result is: {type(sums)}")
print(f"The list of X and Y summed is: {list(sums)}")
```

#### *Question 4.3*

```python
import numpy as np

RPS = input("Enter R, P or S to start playing:")
possible_input = ['R','P','S']

# Text that is printed when the game is over.

Win = "You win, congratulations!"
Lose = "You lost."

# Generates a random number between 0 and 3, then gets the value of that number from the list possible_input

indx = np.random.randint(0,len(possible_input))
result = possible_input[indx]

def RockPapSci(RPS, result):
    """ This function evaluates a random value against user input in Rock-paper-Scissors, then prints if you've won or lost.
        This function written by Klingens on 03-03-2026.

    Parameters
    ----------
    RPS (str) the user input of R, P or S
    result (str) the random vale R, P or S, generated by the programme
    
    Return
    -------
    count (int) the number of times you tried to win
    """
    game = True  # condition
    count = 1
    while game:
        print(f"The programme chose: {result}")
        if RPS != 'R' and 'P' and 'S':  # Evaluates whether the input is valid. If not, enter something again.
            print("Invalid input, try again")
            RPS = input("Enter R, P or S to try again:")
            result = possible_input[indx]
        elif RPS == 'R' and result == 'S':
            game = False  # Ends the game
            print(Win)
        elif RPS == 'P' and result == 'S':
            game = False
            print(Lose)
        elif RPS == 'R' and result == 'P':
            game = False
            print(Lose)
        elif RPS == 'P' and result == 'R':
            game = False
            print(Win)
        elif RPS == 'S' and result == 'P':
            game = False
            print(Win)
        elif RPS == 'S' and result == 'R':
            game = False
            print(Lose)
        else:
            print(f"It's a tie, try again.")
            RPS = input("Enter R, P or S to try again:")  # Asks user to enter a new number due to a draw.
            result = possible_input[indx]
            count += 1  # Ups the count of number of tries by 1
    return count

print(f"You tried {RockPapSci(RPS, result)} times.")
```

#### *Question 4.4*

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

# Create the data.

lengths = np.random.normal(185,8, 100000)
print(f"Set of random lengths: {lengths[:10]}")

# Calculate the mean and standard deviation.

mean = np.mean(lengths)
stdev = np.std(lengths)
print(f"The mean is: {mean} and the standard deviation is {stdev}")

# Probability of finding a value smaller or equal to the indicated height.

ind_height = float(input("Enter a height in cm:"))  # indicated height h
cdf = norm.cdf(ind_height, mean, stdev)  # If you do (h, stdev, mean), it is going to find heights larger than X.
P = cdf*100        # to percentage
print(f"Probability P(X ≤ {ind_height}) is: {P:.2f} %")

# Probability of a height located within 1 standard deviation.

stdev_1 = norm.cdf(mean+stdev, mean, stdev) - norm.cdf(mean-stdev, mean, stdev)
print(f"Probability P(within 1 standard deviation) is: {stdev_1:}")

# Plot the histrogram, mean and standard deviation.

plt.hist(lengths, bins=40, label='Normal distribution')
plt.axvline(mean, color='navy', label=f"Mean = {mean:.2f} cm")
plt.axvspan(mean-stdev, mean+stdev, color='darkturquoise', alpha=0.3, label=f"Standard deviation = {stdev:.2f} cm")

plt.title("Distribution of human lengths")
plt.xlabel("Height ($cm$)")
plt.ylabel("Number of people")

plt.legend()
plt.show()
```
