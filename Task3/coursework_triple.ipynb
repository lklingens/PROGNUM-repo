Write a program which finds all unique Pythagorean triples with values c less than 100. You need a double for loop to iterate over values in a and b. A triple loop is not forbidden by the way.
Include your script and the first 10 triples of the output in your report.

```python
from math import sqrt

def smallc_triples(a, b, c):
    count = 0
    for a in range (1,100):
        for b in range(a,100):
            if count < 10:
                c = sqrt(a**2 + b**2)  # Ensures (a, b, c) is a Pythagorean triple
                if c < 100 and c == int(c):
                    print(f"The triple (a, b, c) is ({a}, {b}, {c})")
                    count += 1
            else:
                return count

print(f"The number of triples shown is: {smallc_triples(0, 0, 0)}")
```
