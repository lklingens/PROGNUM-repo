```python
masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

small_masses = []  # Makes new empty list
print("The length of the masses list is:", len(masses))

# Checks if items in masses are smaller than the moon then adds them to the empty list is they are.

for k in masses:
    if k <= masses[-2]:
        small_masses.append(k)

print("The new list with the masses smaller than the mass of the moon is:", small_masses)  # Prints the new list.

m_slice = slice(6,None,1)  # Slices list from the 6th to last item.
mass_slice = masses[m_slice]  # Slices the masses list.
print("The list with the 5 last masses of the original list is:", mass_slice)  # Prints the new list.

length = len(mass_slice)  #  Length of the slice
mass_sum = sum(mass_slice)  # Sum of the masses in the slice.
avg_mass = mass_sum/(length)  # Calculate average
print("The average mass of the 5 item mass list is:", avg_mass)  # Prints
```

```python
from scipy.constants import G  # Imports G
from math import sqrt

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
object_velocity = {}  # Creates new dictionary

for i in range(len(masses)):  # Execute the loop for the indices in range.
    mass = masses[i]  
    radius = radii[i]
    v_esc = sqrt(2*G*mass/(radius))  # calculates the escape velocity
    print(f"The escape velocity of {names[i]} is {v_esc:e}")  # Prints the escape velocity with the corresponding object name.
    print(f"The mass of {names[i]} is {mass:6.2f}")  # Prints the mass with the corresponding object name
    print()
    object_velocity[names[i]] = v_esc  # Adds the escape velocity to the dictionary with the corresponding object name.

print(object_velocity)
```

```python
from math import atan, degrees

def ang_size():
    D = float(input("Enter a distance:"))
    d = float(input("Enter a size:" ))
    delta = 2*atan(d/(2*D))
    delta_deg = degrees(delta)
    return delta_deg

print(f"The angular size of the object is: {ang_size()}")
```
