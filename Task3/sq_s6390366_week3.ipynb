```python
from scipy.constants import G  # Imports G
from math import sqrt

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
object_velocity = {}  # Creates new dictionary

for i in range(len(masses)):  # Execute the loop for the indices in range.
    mass = masses[i]  
    radius = radii[i]
    v_esc = sqrt(2*G*mass/(radius))  # calculates the escape velocity
    print(f"The escape velocity of {names[i]} is {v_esc:e}")  # Prints the escape velocity with the corresponding object name.
    print(f"The mass of {names[i]} is {mass:6.2f}")  # Prints the mass with the corresponding object name
    print()
    object_velocity[names[i]] = v_esc  # Adds the escape velocity to the dictionary with the corresponding object name.

print(object_velocity)
```

```python
import math

iterations = [10, 100, 500, 1000, 5000, 10000]
sums = []

for n in iterations: 
    # Sums up the equations for each value of n from iterations, then adds the sum to the list called sums.
    s = 0  
    for i in range(1,n):  
        eq = 1/(i**2)  
        s += eq 
    sums.append(s)
    print(s)

print()
converges = True

for i in range(0, len(sums)-1):
    # Calculates the difference between two values of sums, then prints that difference.
    diff = sums[i + 1] - sums[i]
    if diff <= 0: 
        converges = False
    print(diff)

print("The convergence of the given sum is:", converges)
```

```python
from math import atan, degrees

D = float(input("Enter a distance:"))
d = float(input("Enter a size:" ))

def ang_size(d,D):
    """ This function returns two times the arctangent of input parameter distance, D and size, d. The return value is given in degrees.
        This function written by Klingens on 24-02-2026

        Parameters
        -----------
        D (int/float) the input parameter distance,to be multiplied by two; the denominator inside the arctangent,
        d (int/float) the input parameter size; the numerator inside the arctangent,

        Return
        -------
        delta (float) the angular size in degrees.
    """
    delta = 2*atan(d/(2*D))
    delta_deg = degrees(delta)
    return delta_deg

print(f"The angular size of the object is: {ang_size(d, D)}")
```

```python
from math import pi  # To define pi

r = float(input("Enter a radius:"))
h = float(input("Enter a height:"))

def cylinder(r,h):
    """ This function returns the area and the volume of a cylinder, with the parameters r and h
        This function written by Klingens on 24-02-2026

        Parameters
        -----------
        r (int/float) the input parameter radius
        h (int/float) the input parameter height

        Return
        -------
        A (int/float) the area of a cylinder of radius r, height h
        V (int/float) the volume of a cylinder of radius r, height h
    """
    A = 2*pi*r**2 + 2*pi*r*h
    V = pi*r**2*h
    return A,V

A, V = cylinder(r,h)  # Assigns A to the area value and V to the volume value
print(f"The area of the cylinder is {A} and the volume of the cylinder is {V}")

ratio = A/V
print("The ratio of area/volume is:", ratio)
```

```python
import numpy as np

# Asks for input of multiple numbers, splits that into different pieces and maps converts eacht piece into a float. The numbers are then converted to an array.

numbers = map(float, input("Enter you numbers with a space between them:").split())
num_array = np.array(list(numbers))

def stdev(num_array):
    """A function which calculates the standard deviation of an array"""
    mean = sum(num_array)/(len(num_array))
    sigma = np.sqrt(sum((num_array-mean)**2)/(len(num_array)))
    return sigma

def stdev(num_array):
    return np.sqrt(sum((num_array-np.mean(num_array))**2)/(len(num_array)))

print(f"The standard deviation of an array by a defined function is: {stdev(num_array)}")
print(f"The standard deviation of an array by NumPy is: {np.std(num_array)}")
```

```python
import numpy as np
from matplotlib import pyplot as plt
from math import cos, e

x = list(range(-5, 6))
print(x)
y = []

for i in x:
    result = (e**i+e**-i)/(2)
    y.append(result)

print(y)

# Make a plot

plt.plot(x, y, marker='o', color='b', markersize=8, label="Data points")
title = r"Catenary: $\frac{e^{x}+e^{-x}}{2}$"
plt.title(title, fontsize=18, color ='black')
plt.xlabel("x-values", fontsize=12)
plt.ylabel("y-values", fontsize=12)
plt.legend(fontsize=10)
plt.grid()
plt.show()
```
