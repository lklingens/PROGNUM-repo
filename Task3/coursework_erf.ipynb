```python
from math import pi, sqrt, factorial
x = 1

def error_function(x):
    """ This function calculates the error function integral for a value
    This function written by Klingens on 27-02-2026
    
    Parameters
    -----------
    x (int/float) the input parameter number x
    
    Return
    -------
    erf (int/float) the probability in series of trials
    """
    Sums = 0
    k_previous = 0
    for k in range(1000):  # Looping over infinity will overload the kernel, instead we loop over a very large number
        fact = k_previous*k
        if fact == 0:
            fact = 1
        numerator = ((-1)**k)*(x**(2*k+1))
        denominator = (2*k+1)*fact
        fraction = numerator/(denominator)
        Sums += fraction
        k_previous = fact
    erf = (2/(sqrt(pi)))*Sums
    return erf 

print(f"The approximation of the error function for x = 1 is: {error_function(x)}")
```
